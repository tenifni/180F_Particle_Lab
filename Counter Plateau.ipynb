# Single rate

### Counter 1

10s interval

```python
c1_rates=[36,86,150,182, 200, 240,262,323,418,490,541,616,762,750,926,1241,1500,2059,2600,3482,4157,7314]
c1_HV=[1.25,1.30,1.32,1.34,1.35,1.36,1.38,1.40,1.42,1.44,1.45,1.46,1.48,1.50,1.52,1.568,1.60,1.65,1.7,1.75,1.8,1.85]
c1_rates_hz=[i/10 for i in c1_rates]
c1_rates_log=[log(i,10) for i in c1_rates_hz]


fig=plt.scatter(c1_HV, c1_rates_log, color='b')
plt.title('Counter 1')
plt.xlabel('High Voltage (kV)')
plt.ylabel('Rates (Hz) in log scale')
plt.grid("true")
plt.show()
```

# Coincidence

### C1&C3 for C2
C1 and C3 set to 1.6kV. C1,C2,C3 sent to logic unit which is set to Majority function. 
C2 voltage (=HV) is being adjusted

```python
c2_coin_all_rates=[12,57,130,236,306,326,331,343,340,351,351,390,373,352]
c1c3_rates=360
c2_coin_HV=[1.2,1.25,1.3,1.35,1.4,1.45,1.5,1.55,1.60,1.65,1.7,1.75,1.8,1.85]

c2_coin_ratio=[i/360. for i in c2_coin_all_rates]

fig=plt.scatter(c2_coin_HV, c2_coin_ratio, color='r')
plt.title('Counter 2 Coincidence Plateauing with C1 & C3')
plt.xlabel('HV')
plt.ylabel('Ratio of rates')
plt.grid("True")
plt.show()
```
